## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

data = pd.read_csv("FlyRank-Week-1-/data/raw/content_refresh_anonymized.csv")
data_df = pd.DataFrame(data)
data_df['trend_direction'].unique()

# Page Trends

trend_down = data_df.loc[data_df['trend_direction'] == 'down']
trend_up = data_df.loc[data_df['trend_direction'] == 'up']
trend_stable = data_df.loc[data_df['trend_direction'] == 'stable']
trend_flat = data_df.loc[data_df['trend_direction'] == 'flat']
trend_new =  data_df.loc[data_df['trend_direction'] == 'new']

print(f"Trend Down: {len(trend_down)} pages")
print(f"Trend Up: {len(trend_up)} pages")
print(f"Trend Stable: {len(trend_stable)} pages")
print(f"Trend Flat: {len(trend_flat)} pages")
print(f"Trend New: {len(trend_new)} pages|")
print()
print(f"The percentage of pages whose trend is downwards is: {(len(trend_down)/30000)*100:.2f}%")
print(f"The percentage of pages whose trend is upwards is: {(len(trend_up)/30000)*100:.2f}%")
print(f"The percentage of pages whose trend is stable is: {(len(trend_stable)/30000)*100:.2f}%")
print(f"The percentage of pages whose trend is flat is: {(len(trend_flat)/30000)*100:.2f}%")
print(f"The percentage of pages whose trend is new is: {(len(trend_new)/30000)*100:.2f}%")
```
